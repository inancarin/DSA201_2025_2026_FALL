```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
def LinearSearch(myList, val):
    for elem in myList:
        if elem == val:
            return True
    return False
```

```python
np.random.seed(42)
myList = np.arange(1, 1000000, 2) # odd values
myList.shape
```

```python
def BinarySearch(myList, val):
    low, high = 0, len(myList)-1
    while low <= high:
        mid = (low + high) // 2
        if myList[mid] == val:
            return True
        elif myList[mid] > val:
            high = mid - 1
        else:
            low = mid + 1
    return False
```

```python
def BinarySearchRecursive(myList, val, low, high):
    if low > high:
        return False
    mid = (low + high) // 2
    if myList[mid] == val:
        return True
    elif myList[mid] > val:
        return BinarySearchRecursive(myList, val, low, mid-1)
    else:
        return BinarySearchRecursive(myList, val, mid+1, high)
```

```python
for i in range(5, 30):
    maxNum = 2 ** i
    myList = np.arange(1, maxNum, 2) # odd values
    search = 22

    start = time.time()
    res = LinearSearch(myList, search) # O(n)
    end = time.time()
    elapsed = end - start
    print("For n =", maxNum, "time elapsed is", elapsed)
```

```python
for i in range(5, 30):
    maxNum = 2 ** i
    myList = np.arange(1, maxNum, 2) # odd values
    search = 22

    start = time.time()
    res = BinarySearch(myList, search) # O(log n)
    end = time.time()
    elapsed = end - start
    print("For n =", maxNum, "time elapsed is", elapsed)
```

```python
x_values, y_values_linear, y_values_binary, y_values_binaryRec = [], [], [], []
for i in range(5, 30):
    maxNum = 2 ** i
    myList = np.arange(1, maxNum, 2) # odd values
    search = 22

    x_values.append(maxNum)

    start = time.time()
    res = LinearSearch(myList, search) # O(n)
    end = time.time()
    elapsed = end - start
    y_values_linear.append(elapsed)

    start = time.time()
    res = BinarySearch(myList, search) # O(log n)
    end = time.time()
    elapsed = end - start
    y_values_binary.append(elapsed)

    start = time.time()
    res = BinarySearchRecursive(myList, search, 0, len(myList)-1) # O(log n)
    end = time.time()
    elapsed = end - start
    y_values_binaryRec.append(elapsed)

plt.plot(x_values, y_values_linear, label="LinearSearch")
plt.plot(x_values, y_values_binary, label="BinarySearch")
plt.plot(x_values, y_values_binaryRec, label="BinarySearchRec")
plt.legend()
plt.xlabel("Input Size")
plt.ylabel("# of seconds")
plt.show()
```
